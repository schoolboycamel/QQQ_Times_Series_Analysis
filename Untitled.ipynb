```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

df = pd.read_csv('QQQ.csv')

fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
fig.update_layout(
    title='QQQ Historic price',
    yaxis_title='QQQ price',
    shapes = [dict(
        x0='2020-02-20', x1='2020-02-20', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2020-02-20', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='left', text='COVID-19 Pandemic')]
)
fig.show()
```
